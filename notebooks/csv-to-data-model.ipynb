```python
import pandas as pd
import json
```

```python
df = pd.read_csv('../data/lakes_in_austria.csv', delimiter=',')

df[:5]
```

```python
df.columns = ['name', 'state', 'elevation', 'area', 'depth']

df[:5]
```

```python
df['elevation'] = df['elevation'].str.extract(r'(\d+)')
df['area'] = df['area'].str.extract(r'([\d.]+)')
df['depth'] = df['depth'].str.extract(r'(\d+)')

df[:5]
```

```python
entities = []

for index, row in df.iterrows():
    entity = {
        "id": f"urn:ngsi-ld:WaterBody:Lake:{row['name'].replace(' ', '_')}:AT",
        "type": "Lake",
        "name": {
            "type": "Property",
            "value": f"{row['name']}"
        },
        "address": {
            "type": "Property",
            "value": {
                "addressRegion": f"{row['state']}"
            }
        },
        "elevation": {
            "type": "Property",
            "value": float(row['elevation'])
        },
        "area": {
            "type": "Property",
            "value": float(row['area'])
        },
        "depth": {
            "type": "Property",
            "value": float(row['depth'])
        },
        "@context": [
            "https://raw.githubusercontent.com/marvkos/dataModel.WaterBody/refs/heads/main/context.jsonld"
        ]
    }

    entities.append(entity)
```

```python
json.dumps(entities, indent=2)
```

```python
with open('../data/lakes_in_austria.jsonld', 'w') as f:
    f.write(json.dumps(entities, indent=2))
```
